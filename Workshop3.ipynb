```python
import matplotlib.pyplot as plt
import math

# Lists for coordinates
x = []
y = []

# Define interval
start = -4 * math.pi
end = 4 * math.pi
step = 0.1

i = start

while i <= end:
    x.append(i)              # Store x
    y.append(math.sin(i))    # Store sin(x)
    i += step

# Plot graph
plt.plot(x, y, color="red", label="sin(x)")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = sin(x), -4π <= x <= 4π")

plt.grid(True)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import math

x = []   # List for x values
y = []   # List for y values

r = 4
center = [2, 2]

theta_start = 0.0
theta_end = 2 * math.pi
step_size = 0.125

# Number of steps
steps = math.ceil((theta_end - theta_start) / step_size)

theta = theta_start

# Generate points
for i in range(steps):
    x.append(r * math.cos(theta) + center[0])
    y.append(r * math.sin(theta) + center[1])
    theta += step_size

# Plot
plt.figure()

# Plot center point
plt.plot(center[0], center[1], 'ro', label="Center")

# Plot circle
plt.plot(x, y, 'b-', label="Circle")

plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Circle with center (2,2) and radius 4")

# Keep circle shape perfect
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

plt.legend()
plt.show()
```

```python
# importing a library

import matplotlib.pyplot as plt

import math

# Lists to store x and y values

x = []

y= []

# Ellipse parameters

a=3 # semi-minor axis

b=5 # semi-major axis

center =[2,2] # center (p,q)


# Parameter theta

theta_start = 0.0

theta_end = 3*math.pi # make more than 2*pi

step_size = 0.125


# Number of samples

steps = math.ceil((theta_end -theta_start) / step_size)


# initializing theta

theta= theta_start

# calculate x and y cooradinates

for i in range(steps):

    x.append(a*math.cos(theta)+center[0]) # x= a*cos(θ) +p

    y.append(b*math.sin(theta)+center[1]) # y=b*sin(θ) +q

    theta += step_size # incremet the theta by step_size


# Plotting

plt.figure()

# plot the center point

plt.plot(center[0],center[1],'.')

#Plot the ellipse

plt.plot(x,y,'-')

plt.grid()

plt.xlabel("X")

plt.ylabel("Y")

plt.title("Ellipse centered at (2,2) with a=3 , b=5")


# Equal aspect ration to avoid stretching

ax= plt.gca()

ax.set_aspect('equal',adjustable='box')

plt.show()
```

```python
import matplotlib.pyplot as plt

import math

# parabola parameters

q=0 # y-coordinate of vertex

p=0 # x-coordinate of vertex

a=0.5 # controls width and direction

# Step size for x values

step_size = 0.1

# define x range aroung vertex

y_start = p-5

y_end = p + 5

# lists to store coordinates

x_values =[]

y_values =[]

#initialize x
y = y_start

# generate poinst for parabola
while y <= y_end:

    x = 4*a*(y-p)**2+p

    x_values.append(x)

    y_values.append(y)

    y += step_size

#plot the graph

plt.figure()

plt.title(f"Parabola with vertex at ({p},{q}) and a ={a}")

plt.plot(x_values,y_values,'-')

plt.grid()

plt.xlabel("X")

plt.ylabel("Y")

plt.show()
```

# 10. Plot a circle with center at (a, b) and radius r. The values of a=b=2, and r=4, using a while loop.

```python
import matplotlib.pyplot as plt
import math

a, b, r = 2, 2, 4
theta = 0
x_circle, y_circle = [], []

while theta <= 2 * math.pi:
    # Parametric equations: x = r*cos(theta) + a, y = r*sin(theta) + b [cite: 445]
    x_circle.append(r * math.cos(theta) + a)
    y_circle.append(r * math.sin(theta) + b)
    theta += 0.01

plt.plot(x_circle, y_circle)
plt.plot(a, b, 'ro') # Mark the center
plt.gca().set_aspect('equal') # Prevent stretching
plt.title("Circle with center (2,2) and r=4")
plt.grid()
plt.show()
```

# 11. Plot the above circle and the ellipse on the same axes. This is give an idea about the intersection points of the two conic **sections**

```python
import matplotlib.pyplot as plt
import numpy as np

#CIRCLE (from Ex 10)
# Create 100 angles from 0 to 2*pi (full circle)
theta = np.linspace(0, 2*np.pi, 100)


# Circle Formula: x = r * cos(angle) + center_x
# Radius is 4, Center is at (2, 2)
x_c = 4 * np.cos(theta) + 2
y_c = 4 * np.sin(theta) + 2


# ELLIPSE ( from ex 11)
# Define center (p, q) and the two different radii (a, b)

p, q = 2, 2      # Center
a_el = 6         # Horizontal radius (Width)
b_el = 2         # Vertical radius (Height)

# Ellipse Formula: x = horizontal_radius * cos(angle) + center_x
x_e = a_el * np.cos(theta) + p
y_e = b_el * np.sin(theta) + q

# plotting
plt.plot(x_c, y_c, label='Circle')   # Draw circle
plt.plot(x_e, y_e, label='Ellipse')  # Draw ellipse

# IMPORTANT: Forces x and y axes to be same scale
# (Otherwise the circle looks like an oval)
plt.gca().set_aspect('equal')

# Shows the 'Circle' and 'Ellipse' labels
plt.legend()

plt.title("Intersection of Circle and Ellipse")

plt.grid(True)
plt.show()
```

# 12. Plot a Parabola (𝑦 − 𝑝) = 4𝑎(𝑥 − 𝑞)^2 with vertex at (𝑝,𝑞). The values of p, q, and a can be chosen appropriately.

```python
import matplotlib.pyplot as plt
import numpy as np

#define vertex coordinates (p,q) & the width factor 'a'

p = 2
q = 3
a = 0.5

#generate x value around vertex q
# 5 units left & 5 units right of the center

x = np.linspace(q - 5, q + 5, 100)

#calc y using the parabola formula
#formula : y = 4a (x - q)^2 +p
y = 4 * a * (x - q)**2 + p

#plot the curved line
plt.plot(x, y, color='orange', label='Parabola')

#plot a single dot to show the vertex point
plt.plot(q, p, 'ro', label='Vertex (3,2)')

#Add grid & legened for clarity

plt.grid(True)
plt.legend()

plt.show()

```
