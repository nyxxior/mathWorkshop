# b. What are the difference between NumPy array and python list?
Hint:

write the difference on the basis of following

 Data type storage

 Importing modules

 Numerical operations

 Memory consumption

 Fast and convenient to use


# 1. Data type storage:
Python lists can store different data types (strings, ints, floats) together.


NumPy arrays store only one data type (homogeneous), making them more efficient.

# 2. Importing modules:
Python lists are built-in (no import needed).


 NumPy arrays require: import numpy as np.

# 3. Numerical operations:
 Lists don't support direct math (e.g., list * 2 repeats the list).


 NumPy arrays support element-wise math (e.g., array * 2 multiplies every number).

# 4. Memory consumption:
 Python lists use more memory because they store extra information for each element.


 NumPy arrays use significantly less memory as they store data in a compact block.

# 5. Fast and convenient to use:
 Lists are slow for large datasets.


NumPy arrays are much faster and more convenient for scientific and mathematical tasks.

a. Create three vectors in 3D space and plot them using Matplotlib.

U= np.array([1, 0.25 ,0])

V = np.array([0, 1, 0.25])

W = np.array([0.25, 0, 1])

To do 3D plot, you will need to create 3D axes using following code:

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])
origin = np.array([0, 0, 0])

# Create 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vectors using quiver (following Example 7 style for 3D)
ax.quiver(origin[0], origin[1], origin[2], U[0], U[1], U[2], color='r')
ax.quiver(origin[0], origin[1], origin[2], V[0], V[1], V[2], color='b')
ax.quiver(origin[0], origin[1], origin[2], W[0], W[1], W[2], color='g')

# Set limits and labels
ax.set_xlim([0, 1.5])
ax.set_ylim([0, 1.5])
ax.set_zlim([0, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

# b. Create two vectors in 2D space, and add them to create a third vector, and plot all three vectors using Matplotlib.

U = np.array([-0.5, 0.5])

V = np.array([1,1])

The two vectors are added as
W = U + V # adding two vector

```python
import numpy as np
import matplotlib.pyplot as plt

# Create vectors
U = np.array([-0.5, 0.5])
V = np.array([1, 1])
W = U + V  # Vector addition
origin = np.array([0, 0])

fig, ax = plt.subplots()

# Plot using exactly the parameters from Example 7
ax.quiver(origin[0], origin[1], U[0], U[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='g')

ax.set_xlim([-1, 2])
ax.set_ylim([-1, 2])
plt.grid()
plt.show()
```

c. Write a function to calculate magnitude of a vector. For example, calculate the magnitude of the vector be
U = (1.5, 1.5, 2.0).

d. Write a function to calculate the direction cosines of the vector. For example, let the vector be
O = (1.5, 1.5, 2.0).

e. Write a function to calculate Scalar (dot) product of two vectors. For example, let the vectors be
a= (1.5, 1.5, 2.0) and b = (−1.5, 0.5, 2.0).

f. Write a function to calculate Vector (cross) product of two vectors. For example, let the vectors be
 a = (1.5, 1.5, 2.0) and b= (−1.5, 0.5, 2.0)

```python
import numpy as np

# c. Function for Magnitude
def calculate_magnitude(v):
    return np.sqrt(np.sum(v**2))

# d. Function for Direction Cosines
def calculate_direction_cosines(v):
    mag = calculate_magnitude(v)
    return v / mag

# e. Function for Scalar (dot) product
def calculate_dot(a, b):
    return np.dot(a, b)

# f. Function for Vector (cross) product
def calculate_cross(a, b):
    return np.cross(a, b)

# Values from your examples
U_vec = np.array([1.5, 1.5, 2.0])
a_vec = np.array([1.5, 1.5, 2.0])
b_vec = np.array([-1.5, 0.5, 2.0])

# Print results
print("Magnitude:", calculate_magnitude(U_vec))
print("Direction Cosines:", calculate_direction_cosines(U_vec))
print("Dot Product:", calculate_dot(a_vec, b_vec))
print("Cross Product:", calculate_cross(a_vec, b_vec))
```
