#Create a scatter graph of 10 random x and y values between 0 and 50. Label the axis to show
 x – Time taken in seconds
y – Distance reached
Title “Buzzer Game Results”
One you have a working program, add the following code to your program;

,"cyan","magenta"]
And add to your scatter() call the ‘c=colors’ argument like so;

```python
import matplotlib.pyplot as plt
import random

#generate 10 random values betwn 0 & 50
x = [random.randint(0, 50) for i in range(10)]
y = [random.randint(0, 50) for i in range(10)]

#list of color
colors = ["red","green","blue","yellow","pink","orange","purple","brown","cyan","magenta"]

#Scatter plot
plt.scatter(x, y, c=colors)

#labels & titles
plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")

# Show graph
plt.show()
```

 Using the matplotlib.pyplot library, 10 random x and y values between 0 and 50 are generated, plotted, and labeled with titles and axis labels. Adding the c=colors argument to the plt.scatter() function maps a list of 10 distinct colors to each corresponding point, resulting in a scatter plot where each of the 10 data points is rendered in a different color.

```python
import matplotlib.pyplot as plt

labels = ['Bears','Dolphins','Whales']

data = [
    [8,150,80],    # 2017
    [54,77,54],    # 2018
    [93,32,100],   # 2019
    [116,11,76],   # 2020
    [137,6,93],    # 2021
    [184,1,72]     # 2022
]

years = [2017,2018,2019,2020,2021,2022]

# increase figure size (more space)
plt.figure(figsize=(12,8))

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.pie(data[i], labels=labels)
    plt.title(str(years[i]))

# adds spacing between charts
plt.subplots_adjust(wspace=0.6, hspace=0.6)

plt.show()
```
