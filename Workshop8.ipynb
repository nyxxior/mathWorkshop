Consider the following lienear System:

a. x + 3y = 5
   𝑥−2𝑦=−5
   𝑥+𝑦=3

b. 2x = 3y = 3
   4𝑥+6𝑦=6
   3𝑥−2𝑦=1

Solve the linear systems graphically.

```python
# a.
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)

y1 = (5-x)/3
y2 = (x+5)/2
y3 = 3-x

fig, ax = plt.subplots()

ax.plot(x,y1, color='r')
ax.plot(x,y2, color='g')
ax.plot(x,y3, color='b')

ax.text(0.5,0.6, '$x+3y = 5$', color='r')
ax.text(-3,0.5,'$x-2y = -5$', color='g')
ax.text(1.2,2, '$x+y = 3$', color='b')

ax.set_xlim(-4,4)
ax.set_ylim(-2,6)

ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)

ax.set_xticks(list(range(-4,5)))
ax.set_aspect('equal')
ax.grid(True,ls=':')

plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

a. All three lines cross at the same point i.e (1,2).
This means there is one common solution.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)

y1 = (3-2*x)/3
y2 = (6-4*x)/6
y3 = (3*x-1)/2

fig, ax = plt.subplots()

ax.plot(x,y1, color='r')
ax.plot(x,y2, color='g')
ax.plot(x,y3, color='b')

ax.text(0.5,0.6, '$2x+3y = 3$', color='r')
ax.text(-3,0.5,'$4x+6y = 6$', color='g')
ax.text(1.2,2, '$3x-2y = 1$', color='b')

ax.set_xlim(-4,4)
ax.set_ylim(-2,6)

ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)

ax.set_xticks(list(range(-4,5)))
ax.set_aspect('equal')
ax.grid(True,ls=':')

plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

b. Equation (1) and (2) represent the same line, so they overlap completely. Equation (3) is a different line that intersects this line at one point. However, all three lines do not intersect at a single common point, so the system has no solution.

1. Create a matrix 𝑨 of order 5 × 5 with random entries. Calculate the rank of the matrix 𝑨 to check the number of linearly independent columns or rows and to check if the matrix is invertible. What are the other ways to check if the matrix is invertible?

2. Generate a matrix 𝑨 of order 5 × 5 with random entries and generate a vector (column) 𝒙 of length 5 with random entries. Calculate Matrix-vector product 𝒃 = 𝑨𝒙. Now, we have a linear system, i.e. the matrix 𝑨 and a vector 𝒃. Solve this Linear System.

```python
import numpy as np

A = np.random.randint(1,10,(5,5))
x = np.random.randint(1,10,(5,1))

b = np.dot(A, x)

solution = np.linalg.solve(A, b)

print("A:\n", A)
print("\nx:\n", x)
print("\nb:\n", b)
print("\nSolution:\n", solution)
```
