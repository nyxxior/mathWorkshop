# 1. Write a Python function that takes a list and returns a new list with distinct elements from the first list.
# Sample List : [1,2,3,3,3,3,4,5]
## Unique List : [1, 2, 3, 4, 5]

```python
def unique_list(lst):
    unique = []

    for i in lst:
        if i not in unique:
            unique.append(i)

    return unique


sample = [1,2,3,3,3,3,4,5]
print(unique_list(sample))
```

# 2. Write a Python function that takes a number as a parameter and checks whether the number is prime or not.

```python
def check():

  n = int(input("Enter a number"))

  if n<=1:
    return "Not prime"

  for i in range (2,n):
    if n % i == 0:
      return "not prime"

  return "prime"

check()
```

# 3. Write a program to print twin primes less than 1000. If two consecutive odd numbers are both prime then they are known as twin primes

```python
# Function to check if a number is prime
def isPrime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

# Print twin primes less than 1000
for n in range(3, 1000, 2):  # start from 3 and check odd numbers only
    if isPrime(n) and isPrime(n + 2):
        print(f"({n}, {n+2})")
```

# 4. A positive integer is called a perfect number if it is equal to the sum of all of its divisors, including 1 but excluding the number itself.
# For example, 6 = 1 + 2 + 3. Write a function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors (excluding n)

```python
def sumOfDivisors(n):
    divisors = []  # list to store divisors
    for i in range(1, n):  # check numbers from 1 to n-1
        if n % i == 0:     # if i divides n completely
            divisors.append(i)  # add it to the list
    return divisors

num = 15
divs = sumOfDivisors(num)
total = sum(divs)

# Print nicely
print(f"{num} = {' + '.join(str(d) for d in divs)} = {total}")
```

# 5. The smallest perfect number is 6. Write a program that finds and prints out the next perfect number. Hint: You should be able your sumOfDiviors function from your solution to previous question

```python

def sumOfDivisors(n):
    d = []  # create an empty list to store divisors
    for i in range(1, n):  # check all numbers from 1 to n-1
        if n % i == 0:     # if i divides n completely (no remainder)
            d.append(i)  # add i to the list of divisors
    return d

# Start searching for the next perfect number after 6
num = 7  # because 6 is the smallest perfect number

# Keep checking numbers until we find the next perfect number
while True:
    divs = sumOfDivisors(num)  # get divisors of the current number
    total = sum(divs)           # sum up the divisors
    if total == num:

        print(f"The next perfect number is: {num}")
        print(f"{num} = {' + '.join(str(d) for d in divs)} = {total}")
        break
    num += 1
```

# 7. Write a function that converts a decimal number to binary number.

```python
def decimalToBinary(n):
    binary = ""           # start with an empty string
    if n == 0:
        return "0"
    while n > 0:
        binary = str(n % 2) + binary  # get remainder and add in front
        n = n // 2                    # divide number by 2
    return binary

# Eg
num = 14
print("Binary of", num, "is", decimalToBinary(num))
```

# 8. Write a function prodDigits() that inputs a number and returns the product of digits of that number

```python
def prodDigits(n):
    product = 1               # start with 1 because multiplying
    while n > 0:
        digit = n % 10        # get the last digit
        product *= digit      # multiply it to the product
        n = n // 10           # remove the last digit
    return product

# Eg:
num = 555
print("Product of digits of", num, "is", prodDigits(num))
```

# 9. Write a function that will take in an unknown number of arguments and multiply all of them together and run the function for these 2 sets of numbers:
# a. 1,2,3,4,5
# b. 12,13,14

```python
def multiplyAll(*args):
    product = 1                 # start with 1 for multiplication
    for num in args:
        product *= num          # multiply each number
    return product

# Test set a
result_a = multiplyAll(1, 2, 3, 4, 5)
print("Product of 1, 2, 3, 4, 5 is:", result_a)

# Test set b
result_b = multiplyAll(12, 13, 14)
print("Product of 12, 13, 14 is:", result_b)
```
